# Imputing Missing Values

```python
# Imputing missing values

df.isnull().sum() / len(df) * 100
```

```python
# 1 - Dropping the columns having more than 70% of missing values from the dataset

df.drop('deck', axis = 1, inplace = True)
df.drop('Unnamed: 0', axis=1, inplace=True) #Unneccesary column
df.columns
```

```python
df['age'].mean()
df['age'].median()
df['age'].mode()
```

```python
# 2 - Filling out the missing values of column age using mean

df['age'] = df['age'].fillna(df['age'].mean())
df['age']
```

```python
df.isnull().sum()
```

```python
# 3 - Filling out the missing values of embarked & embarked town using mode as they are categorical vals

df['embarked'] = df['embarked'].fillna(df['embarked'].mode()[0])
df['embark_town'] = df['embark_town'].fillna(df['embark_town'].mode()[0])
df['embarked'], df['embark_town']
```

Q1: How many from the age_group has survived? <br>
Q2: How many people are rich who survived? <br>
Q3: The person who survived was alone? <br>
Q4: From which town most of the people alive? <br>

```python
# Q1: How many from the age_group has survived?

df.groupby([ 'survived', 'sex', 'age_group',  ]).size()
```

```python
# Q2: How many people are rich who survived?

df.groupby(['survived', 'who','class', ]).size()

# this shows the real diff, first class travellers survived more than they die comparatively with other class
```

```python
# Q3: The person who survived was alone?

df.groupby(['survived', 'alone', 'sibsp']).size()
```

```python
# Q4: From which town most of the people alive?

df.groupby(['embark_town','survived', ]).size()
# counts per town by survival (0 = died, 1 = survived)
counts = df.groupby(['embark_town', 'survived']).size().unstack(fill_value=0)

# make column names explicit if numeric
counts = counts.rename(columns={0: 'Died', 1: 'Survived'})

# totals and survival rate
counts['Total'] = counts.sum(axis=1)
counts['Survival_Rate'] = counts['Survived'] / counts['Total']

# label whether more survived or died
counts['Comparison'] = counts.apply(lambda r: 'More Survived' if r['Survived'] > r['Died']
                                   else ('More Died' if r['Died'] > r['Survived'] else 'Equal'), axis=1)

# show sorted by survival rate (highest first)
counts_sorted_by_rate = counts.sort_values('Survival_Rate', ascending=False)
counts_sorted_by_rate
```
